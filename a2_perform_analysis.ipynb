```python
from typing import List
import numpy as np
from scipy.stats import norm

def right_sided_cox_stuart_test(data: List[float], alpha: float) -> bool:
    """
    Perform right-sided Cox-Stuart statistical test on the given data.

    Parameters:
    data (List[float]): A list of floats representing the data.
    alpha (float): The significance level.

    Returns:
    bool: True if the null hypothesis is rejected, False otherwise.
    """
    n = len(data)
    x_bar = np.mean(data)
    s = np.std(data, ddof=1)
    z = (x_bar - 0.5) / (s / np.sqrt(n))
    z_alpha = norm.ppf(1 - alpha)
    return z > z_alpha



data = grp_df_count_pytorch['Count'].tolist()
alpha = 0.05

result = right_sided_cox_stuart_test(data, alpha)

if result:
    print("The data is increasing from left to right.")
else:
    print("The data is not increasing from left to right.")
```
